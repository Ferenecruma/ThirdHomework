```python
import pandas as pd

from pure_python import main as pure_python_main
from ctypes_python import main as ctypes_python_main
from numpy_python import main as numpy_python_main


matrix_size = [5, 10, 30, 50, 70, 90, 110, 150, 200]
```

```python
execution_times = [(pure_python_main(size), ctypes_python_main(size), numpy_python_main(size)) for size in matrix_size]
```

```python
# Create pandas dataframe from the list of tuples
# Index of the dataframe should be matrix size
df = pd.DataFrame(execution_times, columns=['Pure Python', 'Ctypes', 'Numpy'], index=matrix_size)
df
```

```python
# Create a line plot with a custom plot size, also add labels for the x and y axis
# For the x axis label should be "Matrix size" and for the y axis label should be "Execution time"
df.plot(figsize=(16, 10), xlabel='Matrix size', ylabel='Execution time')
```

```python
# Create plot only for the Ctypes and Numpy columns
df[['Ctypes', 'Numpy']].plot(figsize=(16, 10), xlabel='Matrix size', ylabel='Execution time')
```
